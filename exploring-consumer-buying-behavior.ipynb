```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/consumers-buying-behavior/social_ads.csv")
```

```python
data.head(2)
```

```python
data.describe()
```

```python
# Visualization 1: Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Histogram of Estimated Salary
plt.figure(figsize=(10, 6))
sns.histplot(data['EstimatedSalary'], bins=20, kde=True, color='salmon')
plt.title('Histogram of Estimated Salary')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 3: Countplot of Purchased
plt.figure(figsize=(8, 8))
ax = sns.countplot(x='Purchased', data=data, palette='pastel')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Countplot of Purchased')
plt.xlabel('Purchased')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Scatter plot of Age vs. Estimated Salary colored by Purchased
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=data, palette='deep')
plt.title('Scatter plot of Age vs. Estimated Salary colored by Purchased')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Visualization 5: Boxplot of Age vs. Purchased
plt.figure(figsize=(8, 5))
sns.boxplot(x='Purchased', y='Age', data=data, palette='coolwarm')
plt.title('Boxplot of Age vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 6: Boxplot of Estimated Salary vs. Purchased
plt.figure(figsize=(8, 5))
sns.boxplot(x='Purchased', y='EstimatedSalary', data=data, palette='viridis')
plt.title('Boxplot of Estimated Salary vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Visualization 7: Violin plot of Age vs. Purchased
plt.figure(figsize=(8, 5))
sns.violinplot(x='Purchased', y='Age', data=data, palette='muted')
plt.title('Violin plot of Age vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 8: Violin plot of Estimated Salary vs. Purchased
plt.figure(figsize=(8, 5))
sns.violinplot(x='Purchased', y='EstimatedSalary', data=data, palette='rocket')
plt.title('Violin plot of Estimated Salary vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Visualization 9: Pairplot of all numerical variables
sns.pairplot(data, hue='Purchased', palette='husl')
plt.suptitle('Pairplot of all numerical variables colored by Purchased', y=1.02)
plt.show()
```

```python
# Visualization 10: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation heatmap')
plt.show()
```
